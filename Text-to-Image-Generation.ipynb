```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
print("GPU available:", torch.cuda.is_available())
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")  # GPU use
```

```python
prompt = "a dreamy night sky full of stars and milky way, glowing lake, mountains silhouette, magical atmosphere, ultra realistic, 4k"

image = pipe(
    prompt,
    guidance_scale=7.5,
    num_inference_steps=40
).images[0]

image
```

```python
image.save("output.png")
```
